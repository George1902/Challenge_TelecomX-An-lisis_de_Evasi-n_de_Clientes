Ahora que has identificado las inconsistencias, es momento de aplicar las correcciones necesarias. Ajusta los datos para asegurarte de que est√©n completos y coherentes, prepar√°ndolos para las siguientes etapas del an√°lisis.

### 1. Normalizaci√≥n de Textos (Strings)
Es com√∫n que en bases de datos de clientes, los valores var√≠en entre may√∫sculas y min√∫sculas o tengan espacios extra. Vamos a estandarizar todo a min√∫sculas y quitar espacios en los extremos.

```python
# Aplicar limpieza a todas las columnas de tipo objeto
for col in datos_clientes.select_dtypes(include=['object']).columns:
    # Convertir a min√∫sculas y eliminar espacios en blanco al inicio/final
    datos_clientes[col] = datos_clientes[col].str.lower().str.strip()

print("Muestra de categor√≠as normalizadas:")
print(datos_clientes['account.Contract'].unique())
```

### 2. Correcci√≥n de la columna Total (Cargos Totales)
Esta es la inconsistencia m√°s cr√≠tica. Los espacios vac√≠os (' ') impiden que la columna sea num√©rica. Usaremos replace y to_numeric.

```python
# 1. Identificar la columna (usualmente conta_cobranza.Total)
col_total = 'account.Charges.Total'

# 2. Reemplazar celdas vac√≠as por NaN usando NumPy
datos_clientes[col_total] = datos_clientes[col_total].replace('', np.nan)

# 3. Forzar el cambio a tipo float
datos_clientes[col_total] = pd.to_numeric(datos_clientes[col_total], errors='coerce')

# 4. Los valores NaN en 'Total' suelen ser clientes con 0 meses de antig√ºedad.
# Vamos a llenar esos nulos con 0.
datos_clientes[col_total] = datos_clientes[col_total].fillna(0)

print(f"Tipo de dato final de {col_total}: {datos_clientes[col_total].dtype}")
```

### 3. Identificaci√≥n de Servicios Espec√≠ficos (startswith / contains)
Imagina que Telecom X quiere agrupar a todos los clientes que usan alg√∫n tipo de servicio de fibra. Podemos usar contains para crear una columna de control.

```python
# Ejemplo de b√∫squeda de patrones: Clientes con Fibra √ìptica
filtro_fibra = datos_clientes['internet.InternetService'].str.contains('fiber', na=False)

# Si quisi√©ramos verificar m√©todos de pago autom√°ticos
datos_clientes['pago_automatico'] = datos_clientes['account.PaymentMethod'].str.contains('automatic|bank', na=False)

print(datos_clientes[['account.PaymentMethod', 'pago_automatico']].head())
```

## Estandarizaci√≥n y transformaci√≥n de datos (opcional)

La estandarizaci√≥n y transformaci√≥n de datos es una etapa opcional, pero altamente recomendada, ya que busca hacer que la informaci√≥n sea m√°s consistente, comprensible y adecuada para el an√°lisis. Durante esta fase, por ejemplo, puedes convertir valores textuales como "S√≠" y "No" en valores binarios (1 y 0), lo que facilita el procesamiento matem√°tico y la aplicaci√≥n de modelos anal√≠ticos.

Adem√°s, traducir o renombrar columnas y datos hace que la informaci√≥n sea m√°s accesible y f√°cil de entender, especialmente cuando se trabaja con fuentes externas o t√©rminos t√©cnicos. Aunque no es un paso obligatorio, puede mejorar significativamente la claridad y comunicaci√≥n de los resultados, facilitando la interpretaci√≥n y evitando confusiones, especialmente al compartir informaci√≥n con stakeholders no t√©cnicos.

### 1. Traducci√≥n de Columnas (Claridad Estrat√©gica)
Vamos a renombrar las columnas clave para que el dataset sea totalmente comprensible.

```python
# Mapeo de nombres de columnas (T√©cnico -> Negocio)
nombres_nuevos = {
    'customer.tenure': 'Meses_Permanencia',
    'internet.InternetService': 'Tipo_Internet',
    'account.Contract': 'Tipo_Contrato',
    'account.PaymentMethod': 'Metodo_Pago',
    'account.Charges.Monthly': 'Gasto_Mensual',
    'account.Charges.Total': 'Gasto_Total',
    'account.Charges.Daily': 'Gasto_Diario',
    'Churn': 'Evasion'
}

datos_clientes.rename(columns=nombres_nuevos, inplace=True)
```

### 2. Binarizaci√≥n de Variables (Transformaci√≥n Matem√°tica)
Convertir "si/no" en "1/0" es fundamental. Usaremos un mapeo sencillo para las columnas binarias.

```python
# Creamos un diccionario de mapeo, ajustado para 'yes', 'no' y cadenas vac√≠as.
mapeo_binario = {'yes': 1, 'no': 0, '': 0}

# Aplicamos a las variables que son de tipo cadena y necesitan binarizaci√≥n.
# 'customer.SeniorCitizen' es num√©rico (0 o 1) por defecto de json_normalize, por lo que no necesita este mapeo.
columns_to_binarize = ['Evasion', 'customer.Partner', 'customer.Dependents']

for col in columns_to_binarize:
    if col in datos_clientes.columns:
        # Llenar posibles None/NaN con cadena vac√≠a antes de operaciones de string
        datos_clientes[col] = datos_clientes[col].fillna('')
        # Asegurar que la columna sea de tipo string y normalizada antes de mapear
        datos_clientes[col] = datos_clientes[col].astype(str).str.lower().str.strip()
        datos_clientes[col] = datos_clientes[col].map(mapeo_binario)

# Asegurar que 'customer.SeniorCitizen' sea num√©rico (0 o 1)
# Es crucial que esta columna no se mapee con mapeo_binario si ya es 0/1 int.
# json_normalize lo trae como int, as√≠ que nos aseguramos que siga as√≠ o se convierta.
if 'customer.SeniorCitizen' in datos_clientes.columns:
    datos_clientes['customer.SeniorCitizen'] = pd.to_numeric(datos_clientes['customer.SeniorCitizen'], errors='coerce')
    datos_clientes['customer.SeniorCitizen'] = datos_clientes['customer.SeniorCitizen'].fillna(0).astype(int)

print(datos_clientes[['Evasion', 'Meses_Permanencia']].head())
```

### 3. Estandarizaci√≥n de Categor√≠as con NumPy
Para las columnas con m√∫ltiples categor√≠as (como el tipo de contrato), es mejor asegurar que los nombres sean cortos y claros.

```python
# Simplificaci√≥n de nombres de contratos usando np.where o replace
datos_clientes['account.Contract'] = datos_clientes['account.Contract'].replace({
    'month-to-month': 'Mensual',
    'one year': 'Anual',
    'two year': 'Bianual'
})

# Verificamos la consistencia final
print("Categor√≠as de Contrato estandarizadas:")
print(datos_clientes['account.Contract'].unique())
```

### 4. Detecci√≥n de Valores At√≠picos (Outliers) con NumPy
Usando la regla del Rango Intercuart√≠lico (IQR), podemos ver si hay clientes con cargos excesivamente altos que podr√≠an estar distorsionando el an√°lisis.

```python
Q1 = datos_clientes['Gasto_Mensual'].quantile(0.25)
Q3 = datos_clientes['Gasto_Mensual'].quantile(0.75)
IQR = Q3 - Q1

limite_superior = Q3 + 1.5 * IQR
outliers = datos_clientes[datos_clientes['Gasto_Mensual'] > limite_superior]

print(f"L√≠mite superior para gastos normales: ${limite_superior:.2f}")
print(f"N√∫mero de clientes con gastos at√≠picos: {len(outliers)}")
```

### 1. Visualizaci√≥n de la Proporci√≥n Global
El primer gr√°fico debe responder a la pregunta: ¬øQu√© tan grave es el problema? Usaremos un gr√°fico de barras para ver la frecuencia y un gr√°fico de pastel para la proporci√≥n

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar datos frescos
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/5adf5f2f87a72bd8b2e854db16e2f2f7c499081a/TelecomX_Data.json'
df_raw = pd.read_json(url)
datos_clientes = pd.json_normalize(df_raw.to_dict(orient='records'))

# 2. Renombrado de columnas esenciales (Ingl√©s -> Espa√±ol)
datos_clientes.rename(columns={
    'customer.tenure': 'Meses_Permanencia',
    'account.Charges.Monthly': 'Gasto_Mensual',
    'Churn': 'Churn_Original'  # Lo guardamos con este nombre para no confundir
}, inplace=True)

# 3. CREAR LA COLUMNA DE EVASI√ìN (Aqu√≠ est√° el truco)
# Verificamos qu√© palabras hay en 'Churn_Original' antes de transformar
print("Valores originales encontrados:", datos_clientes['Churn_Original'].unique())

# Transformaci√≥n robusta: Si dice 'si', 'yes' o '1' es Evasi√≥n (1), de lo contrario es 0.
datos_clientes['Evasion'] = np.where(
    datos_clientes['Churn_Original'].astype(str).str.lower().str.strip().isin(['si', 'yes', '1', 's']),
    1,
    0
)

# 4. Verificaci√≥n del conteo real
conteo = datos_clientes['Evasion'].value_counts()
print("\nNuevo conteo de Evasi√≥n:")
print(conteo)
```

```python
if len(conteo) > 1:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Gr√°fico de Barras con porcentajes
    sns.countplot(x='Evasion', data=datos_clientes, hue='Evasion', palette='magma', ax=ax1, legend=False)
    ax1.set_title('Distribuci√≥n de Evasi√≥n (Frecuencia)', fontsize=14)
    ax1.set_xticklabels(['Permanecen (0)', 'Se Van (1)'])

    total = len(datos_clientes)
    for p in ax1.patches:
        ax1.annotate(f'{100 * p.get_height() / total:.1f}%',
                     (p.get_x() + p.get_width() / 2, p.get_height() + 50), ha='center', fontweight='bold')

    # Gr√°fico de Pastel
    ax2.pie(conteo, labels=['Permanecen', 'Se Van'], autopct='%1.1f%%',
            colors=['#4CAF50', '#FF5722'], startangle=90, explode=[0, 0.1])
    ax2.set_title('Proporci√≥n de Churn', fontsize=14)

    plt.show()
else:
    print("ALERTA: Los datos siguen mostrando una sola categor√≠a. Revisa si el archivo JSON tiene la columna 'Churn' poblada.")
```

### 2. Relaci√≥n entre Evasi√≥n y Gasto Mensual
Ahora que sabemos cu√°ntos se van, veamos si el precio influye. Un gr√°fico de densidad (KDE) nos permite ver si los clientes que se van suelen tener facturas m√°s altas.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n del estilo visual
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Crear el gr√°fico de densidad (KDE)
# Usamos 'hue' para separar a los que se quedan (0) de los que se van (1)
sns.kdeplot(
    data=datos_clientes,
    x='Gasto_Mensual',
    hue='Evasion',
    fill=True,
    common_norm=False,
    palette='magma',
    alpha=0.5,
    linewidth=2
)

# Personalizaci√≥n de t√≠tulos y ejes
plt.title('Influencia del Precio: Densidad de Gasto Mensual por Evasi√≥n', fontsize=15, pad=20)
plt.xlabel('Gasto Mensual ($)', fontsize=12)
plt.ylabel('Densidad de Clientes', fontsize=12)
plt.legend(title='Evasi√≥n', labels=['Se Van (1)', 'Se Quedan (0)'])

plt.show()
```

### Conteo de evasi√≥n por variables num√©ricas

En este paso, explora c√≥mo las variables num√©ricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (evasi√≥n) y los que no cancelaron.

Este an√°lisis ayuda a entender si ciertos valores num√©ricos est√°n m√°s asociados con la evasi√≥n, proporcionando insights sobre los factores que influyen en el comportamiento de los clientes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Asegurar nombres de columnas correctos
# Mapeamos los nombres que el c√≥digo espera a lo que probablemente hay en el DataFrame
mapeo_columnas = {
    'account.Charges.Total': 'Gasto_Total',
    'customer.tenure': 'Meses_Permanencia'
}

for original, nuevo in mapeo_columnas.items():
    if original in datos_clientes.columns and nuevo not in datos_clientes.columns:
        datos_clientes.rename(columns={original: nuevo}, inplace=True)

# 2. Limpieza crucial: Convertir Gasto_Total a num√©rico (a veces viene como texto con espacios)
if 'Gasto_Total' in datos_clientes.columns:
    datos_clientes['Gasto_Total'] = pd.to_numeric(datos_clientes['Gasto_Total'], errors='coerce')
    datos_clientes['Gasto_Total'] = datos_clientes['Gasto_Total'].fillna(0)

# 3. Gr√°ficos de Distribuci√≥n Num√©rica (Boxplots corregidos)
variables_num = ['Meses_Permanencia', 'Gasto_Total']
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for i, var in enumerate(variables_num):
    if var in datos_clientes.columns:
        # Usamos hue='Evasion' y legend=False para evitar el FutureWarning
        sns.boxplot(data=datos_clientes, x='Evasion', y=var, hue='Evasion',
                    palette='magma', ax=axes[i], legend=False)

        axes[i].set_title(f'Distribuci√≥n de {var}', fontsize=14, fontweight='bold')
        axes[i].set_xticks([0, 1])
        axes[i].set_xticklabels(['Permanecen (0)', 'Se Van (1)'])
    else:
        axes[i].set_title(f'Error: {var} no encontrada')

plt.tight_layout()
plt.show()

# 4. Histograma de Permanencia (Curva de Supervivencia)
plt.figure(figsize=(10, 6))
sns.histplot(data=datos_clientes, x='Meses_Permanencia', hue='Evasion',
             kde=True, element="step", palette='magma')
plt.title('Concentraci√≥n de Evasi√≥n por Antig√ºedad', fontsize=15)
plt.xlabel('Meses en la Compa√±√≠a')
plt.ylabel('Cantidad de Clientes')
plt.show()
```
